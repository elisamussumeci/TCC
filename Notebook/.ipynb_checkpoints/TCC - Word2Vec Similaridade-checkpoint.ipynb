# Criação do modelo Word2Vec

```python
# gerador de palavras
def doc_gen(data):
    stopw = stopwords.words('portuguese') +['']
    for doc in data:
        try:
            list_words= nltk.tokenize.wordpunct_tokenize(doc['cleaned_text'])
            ids.append(doc['_id'])
        except KeyError:
            continue
        important_words = [word.lower().strip().strip(punctuation) for word in list_words if word.lower().strip().strip(punctuation) not in stopw]
        yield important_words

```

```python
# corpora gensim com todas as palavras, dictionary[id] retorna a palavra referente ao id
dictionary = corpora.Dictionary(doc_gen(data))
```

```python
# lista de documentos, cada documento é uma lista de tuplas onde documento[0] é uma tupla com id e quantidade da
# palavra no documento
ids=[]
corpus = [dictionary.doc2bow(text) for text in doc_gen(data)]
```

```python
# matriz tfidf, onde cada pos é um documento
tfidf = models.TfidfModel(corpus)
```

```python
len(ids),len(corpus)
```

```python
for i in corpus[:]:
    t=np.array(list(dict(tfidf[i]).values()))
    if not np.isfinite(t).all():
        print(t)
```

```python
#carrega o modelo word2vec para o mediacloud (possui todas as palavras do banco do mediacloud)
model= Word2Vec.load('Models/MediaCloud_w2v')
```

```python
#model.most_similar(positive=['charlie','terrorismo'])
```

```python
#constrói matriz de documentos. Cada linha da matriz é um documento
docvs = np.zeros((len(corpus), model.syn0.shape[1]), dtype=float)
for n,doc in enumerate(corpus):
    ti = dict(tfidf[doc])
    for word in doc:
        w = dictionary[word[0]]
        try:
            word_vector = model[w]
            try:
                assert np.isfinite(word_vector).all()
            except AssertionError:
                print(w)
        except KeyError:
            word_vector = np.zeros((1,model.syn0.shape[1]))
#             print('pulei ', w)
#         word_vector.shape = 1, model.syn0.shape[1]
        docvs[n,:] = docvs[n,:] + word_vector*ti[word[0]]
    
```

##Análise de Clusters

```python
def cluster_vectors(docvs, method='DBS'):
    print("Calculating Clusters.")
    X = docvs
    if method == 'AP':
        af = AffinityPropagation(copy=False).fit(X)
        cluster_centers_indices = af.cluster_centers_indices_
        labels = af.labels_
        n_clusters_ = len(set(labels))
    elif method == 'DBS':
        print("Computing DBSCAN")
        db = DBSCAN(eps=0.03, min_samples=5, algorithm='brute', metric='cosine').fit(X)
        labels = db.labels_
        n_clusters_ = len(set(labels))
    elif method == 'AC':
        print("Computing Agglomerative Clustering")
        ac = AgglomerativeClustering(10).fit(X)
        labels = ac.labels_
        n_clusters_ = ac.n_clusters
    elif method == 'KM':
        print("Computing MiniBatchKmeans clustering")
        km = MiniBatchKMeans(n_clusters=300, batch_size=200).fit(X)
        labels = km.labels_
        n_clusters_ = len(km.cluster_centers_)

    print('Estimated number of clusters: %d' % n_clusters_)
    # print("Homogeneity: %0.3f" % metrics.homogeneity_score(labels_true, labels))
    # print("Completeness: %0.3f" % metrics.completeness_score(labels_true, labels))
    # print("V-measure: %0.3f" % metrics.v_measure_score(labels_true, labels))
    # print("Adjusted Rand Index: %0.3f"
    # % metrics.adjusted_rand_score(labels_true, labels))
    # print("Adjusted Mutual Information: %0.3f"
    #       % metrics.adjusted_mutual_info_score(labels_true, labels))
    # print("Silhouette Coefficient: %0.3f"
    #       % metrics.silhouette_score(X, labels, metric='sqeuclidean'))

    return X, labels
```

```python
X, labels = cluster_vectors(docvs, "DBS")
```

```python
#normaliza toda a matriz docvs
docvs_norm = (docvs / sqrt((docvs ** 2).sum(-1))[..., newaxis]).astype('float')
```

```python
# calcula distancia de todos os os documentos com todos os documentos
dists = dot(docvs_norm, docvs_norm.T)
```

```python
#matriz triangular de distancias cosseno
dists_triu = np.triu(dists,k=1)
```

```python
def consulta(id,data):
    for i in data:
        if i['_id'] == id:
            return i

def create_time_list(data,ids):
    lista = []
    for i in ids:
        article = consulta(i,data)
        if 'published' in article:
             lista.append(article['published'])
        else:
            lista.append(article['updated'])
    return lista

#cria matriz triangular de diferença de tempos
# cada posição x_ij, é a diferença do article da linha i com o article da linha j 
def create_time_dif(time_list):
    times_dif = np.zeros((dists.shape))
    for l,i in enumerate(time_list):
        c = l
        for j in time_list[l:]:
            dif = i-j
            times_dif[l,c] = dif.total_seconds()/3600
            c+=1
    return(times_dif)

```

###Rede de similaridade

```python
from urllib.parse import urlparse

def get_pos(data, pub_i, column_list, time_max, sim_min):
    sim = max(column_list)
    if sim < sim_min:
        return None

    pos = column_list.index(sim)
    time_dif = (data[pos]['published'] - pub_i).total_seconds()/3600
    if time_dif > time_max:
        column_list[pos] = 0
        get_pos(data,pub_i, column_list, time_max, sim_min)
    else:
        return pos

def create_graph(dists_triu, data, time_max=164, sim_min=0.8):
    size = dists_triu.shape[0]
    G = nx.DiGraph()
    G.add_node(0, step=0, date=data[0]['published'], domain=urlparse(data[0]['link']).netloc, _id=data[0]['_id'],
              children=[])
    date_init = data[0]['published']
    for i in range(1,size):
        pub_i = data[i]['published']
        column = list(dists_triu[:,i])
        pos = get_pos(data, pub_i, column, time_max, sim_min)

        if pos != None:
            if pos not in G.nodes():
                domain_1 = urlparse(data[pos]['link']).netloc
                G.add_node(pos, date=data[pos]['published'], domain=domain_1,
                           _id=data[pos]['_id'], children=[])
            if i not in G.nodes():
                domain_2 = urlparse(data[i]['link']).netloc
                G.add_node(i, date=pub_i, domain=domain_2, _id=data[i]['_id'], children=[])

            G.add_edge(pos,i)
    return G
```

```python
def domain_graph(H):
    G = H.copy()
    one_week = 168
    # TODO renomear variavel
    for indice in G.nodes():
        preds = G.predecessors(indice)
        if preds != []:
            pred = G.node[preds[0]]
            node = G.node[indice]
            time_dif = round((node['date']-pred['date']).total_seconds()/3600)
            if node['domain'] == pred['domain'] and time_dif < one_week:
                for suc in G.successors(indice):
                    G.add_edge(preds[0], suc)
                G.node[preds[0]]['children'].append(node['_id'])
                G.remove_node(indice)
    return G
```
